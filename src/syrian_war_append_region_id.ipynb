```python
import pandas as pd
import geopandas as gpd
from shapely.geometry import Point, Polygon
```

```python
shapefile = '../data/syr_admin_json_190321.json'
gdf = gpd.read_file(shapefile)
gdf['id'] = gdf.index
```

```python
df = pd.read_csv('../data/syrian_war.csv')
```

```python
regions = {}
for index in gdf.index:
    regions[index] = Polygon(gdf.at[index, 'geometry'])
```

```python
region_mapping = {}
for index, row in df.iterrows():
    point = Point(row['longitude'], row['latitude'])
    for k, v in regions.items():
        if v.contains(point):
            region_mapping[index] = k
            break
    else:
        region_mapping[index] = -1
```

```python
df['region_id']= pd.Series(region_mapping)
```

```python
df.drop(df[df['region_id'] == -1].index, inplace=True)
```
