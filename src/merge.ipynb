```python
import pandas as pd
```

```python
df1 = pd.read_csv('../data/2010-03-01-2020-05-03-Middle_East-Syria.csv')
df2 = pd.read_csv('../data/2018-04-30-2021-05-08-Middle_East-Syria_actor_based.csv')
```

```python
df1.shape, df2.shape
```

```python
df1.set_index('data_id', inplace=True)
df2.set_index('data_id', inplace=True)
```

```python
df1.shape, df2.shape
```

```python
df1 = df1[df1.index.duplicated()]
df2 = df2[df2.index.duplicated()]
```

```python
df1.shape, df2.shape
```

```python
df1.columns == df2.columns
```

```python
df1.columns
```

```python
df2.tail()
```

```python
df = pd.concat([df1, df2], axis=0).drop_duplicates()
```

```python
df.to_csv('../data/syrian_war_all.csv')
```

```python
for name in ['event_type', 'sub_event_type', 'actor1', 'source_scale']:
    vc = pd.DataFrame(df[name].value_counts())
    vc.columns = ['count']
    vc['name'] = vc.index
    vc.reset_index(inplace=True, drop=True)
    vc.to_csv(f'../data/sw_code_{name}.csv')
    vc_encode = pd.Series(index=vc['name'], data=vc.index)
    vc_encode.to_csv(f'../data/sw_encode_{name}.csv')
    df[name] = df[name].replace(vc_encode)
    print(name, vc.shape)
    print(vc.head())
    print(vc_encode.head())
```
